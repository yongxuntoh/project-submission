Based on what you discovered above, use appropriate methods to re-type incorrectly typed data.
- Define a function that will allow you to convert participation rates to an appropriate numeric type. Use `map` or `apply` to change these columns in each dataframe.

````python
#code
# To remove % signs from ["Participation"] column and to convert them to float data type

def convert_percent(column):
    """
    Convert the percentage string to an actual floating point percent
    - Remove %
    - Divide by 100 to make decimal
    """
    column = column.apply(lambda x: x.replace("%", "")).astype('float') / 100
    return column
    
````

````python
# Run function to convert 
sat_2017["Participation"] = convert_percent(sat_2017["Participation"])
act_2017["Participation"] = convert_percent(act_2017["Participation"])
````

````python
# Define new function to convert participation to float percentage that also handles NaN
def convert_percent2(column):
    """
    Convert the percentage string to an actual floating point percent
    - Remove %
    - Divide by 100 to make decimal
    """
    try: 
        return column.replace('%', '').astype(float) / 100
    except AttributeError:
        return np.NaN
    return column
````

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code

def stdev(list_of_numbers):
    mean = sum(list_of_numbers)/len(list_of_numbers)
    mean_diff_squared = []
    
    for number in list_of_numbers:
        mean_diff_squared.append((mean - number)**2)
        
    return (sum(mean_diff_squared)/(len(list_of_numbers)-1))**0.5
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where:

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, plot_title):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(15,8)) # You'll want to specify your figsize
    plt.suptitle(plot_title, fontsize=20, y=1.05)
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column], bins=10) # feel free to add more settings
        ax[i].set_title("{}" .format(column), y=1.01); #  plot title
        ax[i].set_ylabel("Frequency")
    plt.tight_layout()
````

### Answer: 
<font color="grey">It is evident that those assumptions do not hold for Math, Reading and Rates. <br/>

Central Limit Theorem states that the distribution of sample means approximates a normal distribution as the sample size gets larger (assuming that all samples are identical in size), regardless of population distribution shape (source: https://www.investopedia.com/terms/c/central_limit_theorem.asp). What this means is that even if the population (in this case the overall US Math score, Reading score, or SAT/ACT Participation Rate) does not follow a normal distribution, the means of sample data collected from the population at random will approximate a normal distribution (around the population "true" mean).<br/>
<br/>
In this exercise, the population sample data are not collected at random but rather, first by state and then aggregated together. As such, they are not accurate reflection of the population but rather, just reflection of their own state.<br/>

To observe a normal distribution, we could either:
<ol type="1">
     <li>Collect random sample data within each state, and plot the mean of those sample data against each other</li>
    <li>Collect random sample data across the USA (without any bias to any state), and then plot the mean of those sample data against each other</li></ol></font>
